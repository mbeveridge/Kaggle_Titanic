## Rose vs Jack, or Female vs Male

```python
# Passengers that survived vs passengers that passed away (count & proportion)
survived = train["Survived"].value_counts()
print(survived)
survived_prop = train["Survived"].value_counts(normalize = True)
print(survived_prop)

# Males that survived vs males that passed away (count & proportion)
survived_males = train["Survived"][train["Sex"] == "male"].value_counts()
print(survived_males)
survived_males_prop = train["Survived"][train["Sex"] == "male"].value_counts(normalize = True)
print(survived_males_prop)

# Females that survived vs Females that passed away (count & proportion)
survived_females = train["Survived"][train["Sex"] == "female"].value_counts()
print(survived_females)
survived_females_prop = train["Survived"][train["Sex"] == "female"].value_counts(normalize = True)
print(survived_females_prop)
```

## Does age play a role?

```python
# Create the column Child and assign to 'NaN'
train["Child"] = float('NaN')

# Assign 1 to passengers under 18, 0 to those 18 or older. Print the new column.
train["Child"][train["Age"] < 18] = 1
train["Child"][train["Age"] >= 18] = 0
print(train["Child"])

# Print normalized Survival Rates for passengers under 18
print(train["Survived"][train["Child"] == 1].value_counts(normalize = True))

# Print normalized Survival Rates for passengers 18 or older
print(train["Survived"][train["Child"] == 0].value_counts(normalize = True))
```

# DataCamp :: 2. Predicting with Decision Trees

## Intro to decision trees

```python
import numpy as np
from sklearn import tree
```

## Cleaning and Formatting your Data

```python
# Convert the male and female groups to integer form
train["Sex"][train["Sex"] == "male"] = 0
train["Sex"][train["Sex"] == "female"] = 1
test["Sex"][test["Sex"] == "male"] = 0
test["Sex"][test["Sex"] == "female"] = 1
```

```python
# Impute the Embarked variable ...("Embark" isn't used by model for 'my_solution_one.csv', but was added for 'my_solution_two.csv')
train["Embarked"] = train["Embarked"].fillna("S")
test["Embarked"] = test["Embarked"].fillna("S")

# Convert the Embarked classes to integer form
train["Embarked"][train["Embarked"] == "S"] = 0
train["Embarked"][train["Embarked"] == "C"] = 1
train["Embarked"][train["Embarked"] == "Q"] = 2
test["Embarked"][test["Embarked"] == "S"] = 0
test["Embarked"][test["Embarked"] == "C"] = 1
test["Embarked"][test["Embarked"] == "Q"] = 2
```

```python
#Print the Sex and Embarked columns
print(train["Sex"])
print(train["Embarked"])
```

## Creating your first decision tree

```python
# Print the train data to see the available features
train.reset_index()
```

```python
# Create the target and features numpy arrays: target, features_one
target = train["Survived"].values
features_one = train[["Pclass", "Sex", "Age", "Fare"]].values
```

```python
# Fit your first decision tree: my_tree_one
my_tree_one = tree.DecisionTreeClassifier()
```

```python
my_tree_one = my_tree_one.fit(features_one, target)
```

```python
# Look at the importance and score of the included features
print(my_tree_one.feature_importances_)
print(my_tree_one.score(features_one, target))
```

## Overfitting and how to control it [my_solution_two.csv]

```python
# Create a new array with the added features: features_two
features_two = train[["Pclass","Age","Sex","Fare", "SibSp", "Parch", "Embarked"]].values
```

```python
#Control overfitting by setting "max_depth" to 10 and "min_samples_split" to 5 : my_tree_two
max_depth = 10
min_samples_split = 5
my_tree_two = tree.DecisionTreeClassifier(max_depth = max_depth, min_samples_split = min_samples_split, random_state = 1)
my_tree_two = my_tree_two.fit(features_two, train["Survived"].values)
```

```python
#Print the score of the new decison tree
print(my_tree_two.score(features_two, train["Survived"].values))
```

## Feature-engineering for our Titanic data set [my_solution_three.csv]

```python
# Create train_two with the newly defined feature (family_size = SibSp + Parch + 1)
train_two = train.copy()
train_two["family_size"] = train_two["SibSp"] + train_two["Parch"] + 1
```

```python
# Create a new feature set and add the new feature
features_three = train_two[["Pclass", "Sex", "Age", "Fare", "SibSp", "Parch", "family_size"]].values
```

```python
# Define the tree classifier, then fit the model
my_tree_three = tree.DecisionTreeClassifier()
my_tree_three = my_tree_three.fit(features_three, target)
```

```python
# Print the score of this decision tree
print(my_tree_three.score(features_three, target))
```

# DataCamp :: 3. Improving your predictions through Random Forests

## A Random Forest analysis in Python

```python
# Import the `RandomForestClassifier`
from sklearn.ensemble import RandomForestClassifier
```

```python
# We want the Pclass, Age, Sex, Fare,SibSp, Parch, and Embarked variables
features_forest = train[["Pclass", "Age", "Sex", "Fare", "SibSp", "Parch", "Embarked"]].values
```

```python
# Building and fitting my_forest (with n_estimators set to 100) ...cf. 'tree.DecisionTreeClassifier()'
forest = RandomForestClassifier(max_depth = 10, min_samples_split=2, n_estimators = 100, random_state = 1)
my_forest = forest.fit(features_forest, target)
```

```python
# Print the score of the fitted random forest
print(my_forest.score(features_forest, target))
```

```python
# Compute predictions on our test set features then print the length of the prediction vector
test_features = test[["Pclass", "Age", "Sex", "Fare", "SibSp", "Parch", "Embarked"]].values
pred_forest = my_forest.predict(test_features)
```

```python
print(len(pred_forest))
```

## Interpreting and Comparing

```python
#Request and print the `.feature_importances_` attribute
print(my_tree_two.feature_importances_)
print(my_forest.feature_importances_)
```

```python
#Compute and print the mean accuracy score for both models
print(my_tree_two.score(features_two, target))
print(my_forest.score(features_forest, target))
```
